### NumPy Trigonometric Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.sin(90)
b=np.cos(90)
```

### In radinans

```python
a_rai=np.sin(90*np.pi/180)
b_rai=np.cos(90*np.pi/180)
```

```python
print(a_rai)
print(b_rai)

```

```python
print(np.tan(90*np.pi/180))
```

## Plot a graph 

```python
x = np.arange(0, 3 * np.pi, 0.1) # create x array of angels from range 0 to 
3*3.14
y = np.sin(x) # create y array of sine values of angles from x array
plt.plot(x, y) # plot grah
plt.title(" Graphical Representation of sine function")
plt.xlabel("x axis ")
plt.ylabel("y axis ")
plt.show() # show plotted graph
```
