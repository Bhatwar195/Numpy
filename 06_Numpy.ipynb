### Python NumPy random module

```python
import numpy as np
import random
```

```python
np.random.random(5)
```

```python
np.random.random((3,3))
```

```python
np.random.randint(1,6,(3,3))
```

```python
np.random.seed(10)
np.random.randint(1,6,(2,3,3))
```

```python
# random 2d arr with rand function
np.random.rand(2,3,3)
```

```python
x= [1,2,3,4] # list
choice_from_x = np.random.choice(x) # retun random single item from sequence
print(choice_from_x )
```
