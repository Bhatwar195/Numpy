### Python NumPy String Operations Methods

```python
import numpy as np
```

```python
str1="Machine Learning "
str2="data Science"
```

```python
add_string=np.char.add(str1,str2)
```

```python
add_string
```

```python
mul=np.char.multiply(str1,2)
```

```python
mul
```

```python
cap_string=np.char.capitalize(str2)
```

```python
cap_string
```

```python
low_string=np.char.lower(str1)
```

```python
low_string
```

```python
string=np.char.title(str2)
string
```

```python
string=np.char.center(str1,30,fillchar="*")
string
```

```python
sp_string=np.char.split(str2)
sp_string
```

```python
sp_line=np.char.splitlines(str1+str2)
```

```python
sp_line
```

```python
join_dmy = np.char.join([",", "/"], ["dmy", "dmy"])
print("Joint dmy by , and / = ", join_dmy)
```
