### Slicing concept in numpy


```python
import numpy as np
```

```python
mx_arr=np.arange(1,101).reshape(10,10)
```

```python
mx_arr
```

```python
mx_arr[2,2]
```

```python
mx_arr[3:]
```

```python
mx_arr[1:2]
```

```python
mx_arr[0,0].ndim
```

```python
mx_arr[:,2] #print all column in a row 
```

```python
mx_arr[3:,2:-4] 
```

```python
mx_arr[1:5,1:5]
```

```python
mx_arr.itemsize
```

```python
mx_arr.dtype
```
