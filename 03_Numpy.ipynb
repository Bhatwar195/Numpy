### NumPy array shape Function

```python
import numpy as np
```

```python
arr_1D = np.array([1, 2, 3])
```

```python
arr_1D
```

```python
shape_arr=np.shape(arr_1D)
print("Shape of Array :",shape_arr)
```

```python
arr_2D = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
shape_2D=np.shape(arr_2D)
("Shape of 2D Array :",shape_2D)
```

```python
arr_3D = np.array([[[1, 2, 3], [1, 2, 3], [1, 2, 3]], [[1, 2, 3], [1, 2, 3], 
[1, 2, 3]], [[1, 2, 3], [1, 2, 3], [1, 2, 3]]])
shape_3D=np.shape(arr_3D)
print("Shape of 3D Array :",shape_3D)
```

```python
arr_2D = np.array([[0, 1, 1,4], [1, 0, 1,0], [1, 1, 0,9],[1,3,4,8]]) 
print("2D array : \n", arr_2D) 
```

### NumPy Array Concatenation and Split

```python
arr1=np.arange(1,10).reshape(3,3)
```

```python
arr1
```

```python
arr2=np.arange(11,20).reshape(3,3)
```

```python
arr2
```

```python
Conn_arr=np.concatenate((arr1,arr2))
print(Conn_arr)
```

### Concatenate of arr1 and arr2 with axis


```python
Conn_arr_ax=np.concatenate((arr1,arr2),axis=1)
print(Conn_arr_ax)
```

```python
Conn_arr=np.concatenate((arr1,arr2),axis=0)
print(Conn_arr)
```
